```python
import torch
from transformers import CLIPTokenizer, CLIPTextModel
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
model.scheduler = DPMSolverMultistepScheduler.from_config(model.scheduler.config)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
text_encoder = CLIPTextModel.from_pretrained("openai/clip-vit-base-patch32")

def generate_images(prompt, batch_size=3):
    try:
        
        text_inputs = tokenizer(prompt, padding="max_length", max_length=77, return_tensors="pt")
        text_inputs = {k: v.to(device) for k, v in text_inputs.items()}  
        
    
        text_encoder.to(device)
        text_embeddings = text_encoder(text_inputs['input_ids'])[0].to(device)
        
        
        images = model(prompt, num_images_per_prompt=batch_size).images
        return images
    except Exception as e:
        print(f"Error generating images: {e}")
        raise

def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        
        img = images[i]
        if img.mode != 'RGB':
            img = img.convert('RGB')
        plt.imshow(img)
        plt.axis("off")
    plt.show()

def on_button_click(b):
    prompt = text_input.value
    images = generate_images(prompt, batch_size=3)
    if images:
        plot_images(images)

text_input = widgets.Text(
    value='photograph of an astronaut riding a horse',
    placeholder='Type your prompt here',
    description='Prompt:',
    disabled=False
)

button = widgets.Button(
    description='Generate Image',
    disabled=False,
    button_style='', 
    tooltip='Click to generate image',
    icon='check' 
)

button.on_click(on_button_click)

display(text_input)
display(button)
```
